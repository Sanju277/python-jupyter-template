### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11, 1)  # Values from 0 to 10
y = x**2  # Squaring each x value

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Customizing the plot
plt.title('Square Function Plot')  # Add title
plt.xlabel('X values')  # X-axis label
plt.ylabel('Y values (x²)')  # Y-axis label
plt.grid(True)  # Add grid
plt.legend()  # Add legend

# Show the plot
plt.show()
```
